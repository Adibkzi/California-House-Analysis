# Project Overview:
In this case study, we will analyze the California housing dataset to build a predictive model for house prices. We will use Python data science libraries such as seaborn, numpy, matplotlib, and pandas for exploratory data analysis (EDA) and visualization. The goal is to understand the factors that influence house prices and develop a predictive model that can estimate the prices accurate.

```python
#Import the Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the dataset
data = pd.read_csv("housing.csv")
data.head()
```

```python
# Data Cleaning
# Drop any rows with missing values
data = data.dropna()
print(data)
```

```python
# Exploratory Data Analysis 
# Question 1: How is the median house price distributed across different districts in California?

# Visualize the distribution of median house prices
plt.figure(figsize=(8, 6))
sns.histplot(data['median_house_value'], kde=True)
plt.title('Distribution of Median House Prices')
plt.xlabel('Median House Price')
plt.ylabel('Frequency')
plt.show()
```

```python

# 2. What are the key factors that influence house prices?

# Convert categorical variable "ocean_proximity" to numerical using one-hot encoding
data_encoded = pd.get_dummies(data, columns=['ocean_proximity'])

# Compute the correlation matrix
corr_matrix = data_encoded.corr()

# Sort the correlation values with respect to median house value
corr_values = corr_matrix['median_house_value'].sort_values(ascending=False)

# Select the top 5 features with the highest correlation
top_features = corr_values[1:6]  # Exclude the target variable itself

# Visualize the correlations using a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=top_features.values, y=top_features.index, orient='h')
plt.title('Top 5 Features Influencing House Prices')
plt.xlabel('Correlation')
plt.ylabel('Feature')
plt.show()
```

```python

# Question 3: Is there a correlation between the house price and the average number of rooms per dwelling?

# Scatter plot of median house value vs. average number of rooms
plt.figure(figsize=(8,6))
sns.scatterplot(x='total_rooms', y='median_house_value',data=data)
plt.title("Correlation between the house price and the average number of rooms")
plt.xlabel("Average Number of Rooms")
plt.ylabel("Median House Value")
plt.show()
```

```python
# 4. How does the location affect house prices in California?
plt.figure(figsize=(8,6))
sns.scatterplot(x='ocean_proximity', y='median_house_value',data=data)
plt.title("Location Affects House Prices")
plt.xlabel("Ocean Proximity")
plt.ylabel("Median House Value")
plt.show()
```
